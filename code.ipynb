```python
# Replace commas with dots and convert to numeric
columns_to_fix = ['CO(GT)', 'C6H6(GT)', 'T', 'RH', 'AH']
for col in columns_to_fix:
    data[col] = data[col].str.replace(',', '.').astype(float)

# Drop irrelevant columns
data = data.drop(columns=['Unnamed: 15', 'Unnamed: 16'])

# Verify the changes
print(data.dtypes)
print(data.head())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Plot boxplots for each feature to visualize the outliers
plt.figure(figsize=(12, 8))
for i, column in enumerate(data_cleaned.columns):
    plt.subplot(4, 3, i + 1)  # Adjust the grid size based on the number of features
    sns.boxplot(data=data_cleaned[column])
    plt.title(f"Boxplot for {column}")
    plt.tight_layout()
plt.show()
```

```python
# Calculate the IQR
Q1 = data_cleaned.quantile(0.25)
Q3 = data_cleaned.quantile(0.75)
IQR = Q3 - Q1

# Filter out the outliers
data_no_outliers = data_cleaned[~((data_cleaned < (Q1 - 1.5 * IQR)) | (data_cleaned > (Q3 + 1.5 * IQR))).any(axis=1)]

# Verify if outliers are removed
print("Data after removing outliers:\n", data_no_outliers)
```

```python
# Define features and target variable
X = data_cleaned.drop(columns=["CO(GT)"]) # Drop target column, you can choose any target column
Y = data_cleaned["CO(GT)"]  # Target variable, e.g., CO(GT)
```

```python
from sklearn.model_selection import train_test_split 
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
# Initialize and train the Random Forest model
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)
rf_model.fit(X_train, Y_train)

# Predict on the test data
y_rf_pred = rf_model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score

# Calculate MSE and R-squared for Random Forest model
mse_rf = mean_squared_error(Y_test, y_rf_pred)
r2_rf = r2_score(Y_test, y_rf_pred)

print(f"Random Forest Mean Squared Error: {mse_rf}")
print(f"Random Forest R-squared: {r2_rf}")
```

```python
import pandas as pd

# Assuming you already have the dataset 'data'

# Define a threshold for NO2(GT) to classify pollution status
# You can adjust the threshold based on domain knowledge or data distribution
pollution_threshold = data['NO2(GT)'].median()

# Create a new column 'pollution_status' based on the threshold
data['pollution_status'] = data['NO2(GT)'].apply(lambda x: 'high' if x > pollution_threshold else 'low')

# Print the first few rows to verify
print(data[['NO2(GT)', 'pollution_status']].head(10))
```

```python
# Define the input data with the same columns used during training
input_data = pd.DataFrame({
    'PT08.S1(CO)': [1.0],   # Example value for 'PT08.S1(CO)'
    'C6H6(GT)': [5.0],      # Example value for 'C6H6(GT)'
    'PT08.S2(NMHC)': [10.0], # Example value for 'PT08.S2(NMHC)'
    'NOx(GT)': [50.0],      # Example value for 'NOx(GT)'
    'PT08.S3(NOx)': [100.0], # Example value for 'PT08.S3(NOx)'
    'NO2(GT)': [25.0],      # Example value for 'NO2(GT)'
    'PT08.S4(NO2)': [200.0], # Example value for 'PT08.S4(NO2)'
    'PT08.S5(O3)': [300.0],  # Example value for 'PT08.S5(O3)'
    'T': [23.0],            # Example value for temperature ('T')
    'RH': [55.0],           # Example value for relative humidity ('RH')
    'AH': [0.7]             # Example value for absolute humidity ('AH')
})

# Make prediction using the trained model
pollution_status_pred = rf_model.predict(input_data)

# Define thresholds for pollution levels
def get_pollution_level(value):
    if value < 10:
        return "Low Pollution"
    elif 10 <= value < 50:
        return "Moderate Pollution"
    else:
        return "High Pollution"

# Output the predicted value and its corresponding pollution level
pollution_level = get_pollution_level(pollution_status_pred[0])
print(f"Predicted Pollution Status: {pollution_level} ({pollution_status_pred[0]:.2f})")
```
